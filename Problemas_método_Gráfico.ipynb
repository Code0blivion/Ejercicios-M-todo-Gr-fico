```python
import numpy as np
import sympy as sym
from matplotlib import pyplot

#Luis David Bautista Pérez
#Cod. 20202020144
#Investigación de Operaciones I
#Problemas propuestos


#A continuación se imprime el planteamiento del problema
print("1. Hallar y gráficar la región solucion de la siguiente desigualdad")
print("2x + y < 5")

def f1(x):
  return 5-2*x

x = range (0,4)

pyplot.plot(x, [f1(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,[f1(i) for i in x], color='red')
pyplot.show()

print("La región en rojo es la region solucion sin contar la recta. Ignorar el pedazo de grafica desde x=2.5, es un error de impresion")
print()
print()

print("2. Hallar y gráficar la región solucion de la siguiente desigualdad")
print("y =< 5")

def f2(x):
  return 5

x = range (0,8)

pyplot.plot(x, [f2(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,[f2(i) for i in x], color='red')
pyplot.show()

print("La región en rojo es la region solucion junto a la recta.")
print()
print()

print("3. Hallar y gráficar la región solucion de la siguiente desigualdad")
print("2(2x - y) < 2(x + y) - 4")

def f3(x):
  return (x/2)+1

x = range (-2,2)

pyplot.plot(x, [f3(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,[f3(i) for i in x],5, color='red')
pyplot.show()
print("La región en rojo es la region solucion sin la recta.")
print()
print()

print("4. Hallar y gráficar la región solucion de las siguientes desigualdades")
print("2x + y > 3")
print("2y - 1 > 0")
print("x >= y")

def f41(x):
  return 3-2*x

def f42(x):
  return 1/2

def f43(x):
  return x

x = range (1,5)

pyplot.plot(range(0,5), [f41(i) for i in range(0,5)])
pyplot.plot(range(0,5), [f42(i) for i in range(0,5)])
pyplot.plot(range(0,5), [f43(i) for i in range(0,5)])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,[f42(i) for i in x],[np.maximum(f41(i),f43(i)) for i in x],5, color='red')
pyplot.show()
print("La región en rojo es la region solucion junto la recta y = x.")
print()
print()

print("5. Hallar y gráficar la región solucion de las siguientes desigualdades")
print("2x + 3y =< 60")
print("x >= 0")
print("y >= 0")

def f2(x):
  return 20 - (2/3)*x

x = range (0,31)

pyplot.plot(x, [f2(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,[f2(i) for i in x], color='red')
pyplot.show()

print("La región en rojo es la region solucion junto a la recta.")
print()
print()

print("6. Maximizar P = 4x + 6y")
print("Sujeto a:")
print("2x + y =< 180")
print("x + 2y =< 160")
print("x + y =< 100")

def f61(x):
  return 180 -2*x

def f62(x):
  return 80-x/2

def f63(x):
  return 100-x

  

x = range (0,95)

pyplot.plot(x, [f61(i) for i in x],x, [f62(i) for i in x],x, [f63(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,np.minimum([f62(i) for i in x],[f63(i) for i in x]), color='red')
pyplot.show()

print("La region en rojo es la region solucion junto a todas las rectas.")


print("La coordenada que maximiza el P es (40, 60) con un valor de 520")
print()
print()


print("EJERCICIO DE APLICACION DE PRODUCTOS X - Y")
print("Maximizar P = 200x + 100y")
print("Sujeto a:")
print("5x + 3y =< 105")
print("2x + 4y =< 70")
print("x >= 0")
print("y >= 0")

def fa1(x):
  return 35 - (5/3)*x

def fa2(x):
  return (35/2)-x/2

  

x = range (0,22)

pyplot.plot(x, [fa1(i) for i in x],x, [fa2(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,np.minimum([fa1(i) for i in x],[fa2(i) for i in x]), color='red')
pyplot.show()

print("La región en rojo es la region solucion junto a todas las rectas.")
print("La coordenada que maximiza el P es (21,0) con un valor de 4200")
print()
print()


```

```python
import numpy as np
import sympy as sym
from matplotlib import pyplot

#Luis David Bautista Pérez
#Cod. 20202020144
#Investigación de Operaciones I
#Problemas practicos - de aplicacion de maximizacion y minimizacion


#Los enunciados de todos los problemas se encuentran en el PDF en el que se encuentran las soluciones a mano

print("Problema de maximización #1")

print("Maximizar P = 7000x + 8000y")
print("Sujeto a:")
print("2x + y =< 19")
print("x + y =< 14")
print("x + 2y =< 20")
print("x >= 0")
print("y >= 0")

def fa1(x):
  return 19-2*x

def fa2(x):
  return 14-x

def fa3(x):
  return 10-x/2

  

x = range (0,11)

pyplot.plot(x, [fa1(i) for i in x],x, [fa2(i) for i in x],x, [fa3(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,np.minimum([fa1(i) for i in x],[fa3(i) for i in x]), color='red')
pyplot.show()

print("La región en rojo es la region solucion junto a todas las rectas.")
print("La coordenada que maximiza el P es (6,7) con un valor maximo de beneficios de 98000 pesetas")
print()
print()


print("Problema de maximización #2")

print("Maximizar P = 8000x + 6000y")
print("Sujeto a:")
print("4x + 2y =< 60")
print("2x + 4y =< 48")
print("6x + 2y =< 76")
print("x >= 0")
print("y >= 0")

def fa1(x):
  return 30-2*x

def fa2(x):
  return 12-x/2

def fa3(x):
  return 38-3*x

  

x = range (0,14)

pyplot.plot(x, [fa1(i) for i in x],x, [fa2(i) for i in x],x, [fa3(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.fill_between(x,np.minimum([fa2(i) for i in x],[fa3(i) for i in x]), color='red')
pyplot.show()

print("La región en rojo es la region solucion junto a todas las rectas.")
print("La coordenada que maximiza el P es (10.4,6.8) con un valor maximo de beneficios de 124000 pesetas")
print()
print()



print("Problema de minimizacion #1")

print("Minimizar P = 400x + 1000y")
print("Sujeto a:")
print("6x + 10y = 1000")
print("x =< 80")
print("y =< 80")
print("x >= 0")
print("y >= 0")

def fa1(x):
  return 100-(3/5)*x

def fa2(x):
  return 80 

x = range (0,81)

pyplot.plot(range(0,34), [fa1(i) for i in range(0,34)],range(33,80), [fa1(i) for i in range(33,80)],x, [fa2(i) for i in x],80,0 )
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.show()

print("La región solución es toda la recta naranja.")
print("La coordenada que minimiza el P es (80,52) con un valor minimo de coste de 84000 pesetas")
print()
print()

print("Problema de minimizacion #2")

print("Minimizar P = 3x + 2y")
print("Sujeto a:")
print("5x + y >= 10")
print("2x + 2y = 12")
print("x + 4y >= 12")
print("x >= 0")
print("y >= 0")

def fa1(x):
  return 10-5*x

def fa2(x):
  return 6-x

def fa3(x):
  return 3-x/4

x = range (0,7)

pyplot.plot(x, [fa1(i) for i in x],x, [fa1(i) for i in x],x, [fa3(i) for i in x],x,[fa2(i) for i in x],range(1,5),[fa2(i) for i in range(1,5)])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.show()

print("La región solución es toda la recta morada.")
print("La coordenada que minimiza el P es (1,5) con un valor minimo de 13 unidades de energia")
print()
print()


```
